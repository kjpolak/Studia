```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import queue
import operator
from functools import reduce
import sys
import time
```

```python
points = np.random.rand(150,2)
```

# S-hull

```python
from scipy.spatial import Delaunay
start = time.time()
s_hull = Delaunay(points)
end = time.time()
print(end - start)
```

```python
plt.figure(figsize=(15,20))
plt.triplot(points[:,0], points[:,1], s_hull.simplices.copy())
plt.plot(points[:,0], points[:,1], 'o', color="red")
plt.show()
```

#  Bowyer-Watson (Incremental)

```python
from incremental.delaunay2D import Delaunay2D
```

```python
start = time.time()

incremental = Delaunay2D()
for point in points:
    incremental.addPoint(point)
    
end = time.time()
print(end - start)
```

```python
plt.figure(figsize=(15,20))
plt.triplot(points[:,0], points[:,1], incremental.exportTriangles().copy())
plt.plot(points[:,0], points[:,1], 'o', color="red")
plt.show()
```
